```python
import json
import os
```

```python
os.chdir(os.getcwd() + "/data")
```

```python
files = os.listdir(os.getcwd())[2:-1]
```

```python
total = len(files)
count = 0
```

```python
for file in files:
    with open(file) as f:
        try:
            data = json.load(f)
        except ValueError:
            os.remove(file)
            continue
    if data['comments'] == []:
        os.remove(file)
    else:
        count += 1
        print file
print float(count)/total
```
